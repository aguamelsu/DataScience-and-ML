```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sy
```

```python
def line_function(x:float)->float:
  return x + 2
```

```python
x = np.linspace(-5,5,100)
plt.plot(x, line_function(x), label="line_funtion")
plt.axhline(y=0, color="r")
plt.axvline(x=0, color="r")
plt.grid(True)
plt.show()
```

```python
def f(x:float) -> float:
  return 2*x + 2 #x=0,y=2 / x=2,y=6
```

```python
x = np.linspace(-5,5,100)
plt.plot(x, f(x), label="line_funtion")
plt.axhline(y=0, color="r")
plt.axvline(x=0, color="r")
plt.legend()
plt.grid(True)
plt.show()
```

```python
def g(x:float) -> float:
  return x**2 + 1 #Cuadratic function
```

```python
x = np.linspace(-5,5,100)
plt.plot(x, g(x), label="line_funtion")
plt.axhline(y=0, color="r")
plt.axvline(x=0, color="r")
plt.plot(0,g(0), 'go')
plt.plot(2,g(2), 'go')

plt.legend()
plt.grid(True)
plt.show()
```

```python
def k(x:float) -> float:
  return x*x - 2*x + 4
```

```python
def dk(x:float) -> float:
  h = 0.000001
  return (k(x+h) - k(x))/h
```

```python
x = np.linspace(-5,5,100)
plt.plot(x, k(x), label="line_funtion")
plt.plot(x, dk(x), label="Derivate line_funtion")
plt.plot(0,dk(0), "ro")
plt.plot(2,dk(2), "ro")
plt.plot(1, dk(1), "ko")

plt.axhline(y=3, color="k")
plt.axhline(y=0, color="g")
plt.axvline(x=0, color="g")


plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def square(x:float)->float:
  return x**2              # ** es a la potencia n

plt.plot(x, square(x))
plt.axhline(y=0, color='r')
plt.axvline(x=0, color='r')
plt.grid(True)
plt.title('Gráfica función cuadrática en el intervalo [0, 4]')
plt.show()
```

```python
x = np.linspace(-10, 10, 1000)
x_2 = np.linspace(-3, 3, 1000)

def sin(x:float)->float:
  return np.sin(x)

def exp(x:float)->float:
  return np.power(2, x)

plt.plot(x, sin(x))
plt.plot(x_2, exp(x_2))
plt.axhline(y=0, color='r')
plt.axvline(x=0, color='r')
plt.grid(True)
plt.title('Graph of sine function in the interval [-10, 10]')
plt.show()
```

 ### **Random Variables**

 - **Discrete random variables**: Can take a countable, finite number of values. *E.g. Heads or tails*. 
  - **Continuos random variables**: Can take any value in a continuos interval. *E.g. An average*

Categorizing variables is important because each random variable is associated with a **distribution function**.

### **⚡Key Ideas...**
The Cumulative Distribution Function (CDF) of a random variable ***X*** is:

$$F(x) = P(X \leq x)$$

$\rightarrow$ This is the probability that *X* is less or equal to a certain value. It's like the **accumulated probability** up to *x*

But what if it was the other way around? Let's say:

$$P(X \geq x)$$

$\rightarrow$ instead of summing probabilities forever like:

$$P(X \geq 3) = P(X=3) + P(X=4) + ... + P(X= \infty)$$

we can use the **complement rule**:

$$P(X\geq k) = 1 - P(X\leq (k-1)) \rightarrow (discrete)$$
$$P(X\geq a) = 1 - P(X\leq a) \rightarrow (continuous)$$

### Bernoulli's distribution

**Bernoulli's** model is applied to a random variable that can only have two values. For this model, we have:

$$E(X)=p$$
$$V(X)=pq$$

Where: 

- *p* is the probability of success
- *q* is the probability of failure
- $E(X)$ is the expected of the random variable
- $V(X)$ is the variance of the random variable.

### Binomial distribution

The binomial distribution can be considered as the sum of *n* independent Bernoulli variables, where the following conditions must be met:
1. There must be a fixed number of statistically independent repeated tests.
2. Each test must be a Bernoulli variable (result in a success or a failure)
3. All the tests must have identical success probabilities.

The binomial model is discrete and has the following:
$$E(X) = np$$
$$V(X) = npq$$

Where:

- *n* is the number of tests
- *p* is the probability of success
- *q* is the probability of failure

*Problem 1*: 

A coin is throwed 10 times. What's the probability of getting 8 or more heads? 

```python
n = 10 # Number of cases
p = 0.5 # Probability of favoural cases (heads)
prob = 1 - binom.cdf(7, n, p) # Probability of getting 8 or more heads 

print(f'The probability of getting 8 or more heads is {round(prob * 100, 2)}%')
```

*Problem 2*: 

It is well known that 90% of the workers in an industrial city are members of a union. In a study of laboural situation, an economist randomly chooses a sample of 15 workers for an interview. What is the probability that the sample contains 12 or less members of the union?

```python
p = 0.9
n = 15
prob = binom.cdf(12, n, p)
print(f'The probability of getting 12 or less workers from the union is: {round(prob * 100, 2)}%')
```

### Poisson Distribution
A lot of experiments do not occur as a defined result of a test, instead they show as points in a randomly set of time, space or volume. Poisson's model is perfect for these cases if:
1. The number of occurences is independent of the unit
2. The expected value of a variable is proportional to the size of the specified unit
3. The probability of more than one occurence in a tiny little specific unit can be rejected

For this model we have:
$$E(X)=\lambda$$
$$V(X)=\lambda$$

Where $\lambda$ is the number of occurences per specified unit. (Poissons distribution is discrete)

*Problem 3*: 

A callcenter receives 4 calls per minute ($\lambda$ = 4). What is the probability that they do NOT receive more than 2 calls in the next 2 minutes?

```python
n = 2
plambda = 8
prob = poisson.cdf(n, plambda)
print(f'The probability of NOT receiving more than 2 calls in the next two minutes is: {round(prob * 100, 2)}%')
```

*Problem 3.1*: 

What is the probability that the call center receives 4 or more calls?

```python
n = 3
plambda = 8
prob = 1 - poisson.cdf(3, 8)
print(f'The probability of receiving 4 or more calls is: {round(prob * 100, 2)}%')
```

*Problem 4*: 

A fabric manufacturer claims that the average number of defects in his product is 1 per 2 squared yard. A squared yard was extracted randomly and it had 3 defects. What is the probability of getting 3 or more defects in whichever squared yard if the manufacturer's claim were true?

```python
plambda = 0.5
n = 3
prob = 1 - poisson.cdf(n, plambda)
print(f'The probability of getting 3 or more deffects is: {round(prob * 100, 2)}%')
```

### **Normal Distribution**
This model is frequently used in economical and commercial analysis. A normal distribution can be expressed in the general and standarized form. 

A random variable *X* has a normal distribution if: 
1. Its a **continuous** random variable
2. Its domain $D = (- \infty, \infty)$ 
3. Can be modeled by its mean $\mu$ and its standard deviation $\sigma$

```python
x = np.linspace(-5, 5, 1000)
pdf = norm.pdf(x) # Probability density function

plt.plot(x, pdf, c="tomato")
plt.title("PDF Normal Distribution")
plt.xlabel("Value")
plt.ylabel("Density")
plt.grid(True)
plt.show()
```

```python
x = np.linspace(-5, 5, 1000)
cdf = norm.cdf(x) # Cumulative distribution Function

plt.plot(x, cdf, c="tomato")
plt.title("CDF Normal Distribution")
plt.xlabel("Value")
plt.ylabel("Probability")
plt.grid(True)
plt.show()
```

```python
n = 10
p = 0.5
x = range(n+1)

pmf = binom.pmf(x, n, p)
plt.stem(x, pmf)
plt.grid(True)
plt.show()
```

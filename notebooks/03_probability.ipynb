```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
import random
from fractions import Fraction
```

1. Cuál es la probabilidad de obtener un número impar en un lanzamiento de dados?

```python
# Rolling a dice
sample_space = {1, 2, 3, 4, 5, 6}
event = {1, 3, 5} # Probability of odd numbers
intersection = len(set.intersection(sample_space, event))

# A probability is defined as the cardinality of the intersection between
# the event and the sample space, over the cardinality of the sample
# space
Fraction(intersection, len(sample_space))
```

2. Cuál es la probabilidad de obtener un número primo o un número par en un lanzamiento de dados?

```python
sample_space = {1, 2, 3, 4, 5, 6}
odd = {2, 4, 6}
prime = {2, 3, 5}
events = prime.union(odd)

events_cardinality = len(set.intersection(events, sample_space))
sample_cardinality = len(sample_space)

Fraction(events_cardinality, sample_cardinality)
```

3. Cuál es la probabilidad de obtener un número par y primo en un lanzamiento de dados?

```python
sample_space = {1, 2, 3, 4, 5, 6}
even = {2, 4, 6}
prime = {2, 3, 5}

events_cardinality = len(set.intersection(even, prime))
sample_cardinality = len(sample_space)

probability = Fraction(events_cardinality, sample_cardinality)

probability
```

```python
def pr(event:set, space:set) -> Fraction:
    return Fraction(len(set.intersection(event, space)), len(space))
```

```python
pr({2, 4, 6}, {1, 2, 3, 4, 5, 6})
```

4. El método de Montecarlo, permite aproximar una probabilidad con base a la repetición continua de un evento. ¿Cuál es la probabilidad de obtener un número par en un lanzamiento de dados?

```python
events = [2, 4, 6]
sample_space = [1, 2, 3, 4, 5, 6]
frec = []
trials = list(range(1, 1001))

for trial in trials:
    total = 0
    for i in range(trial):
        if random.choice(sample_space) in events:
            total += 1
    frec.append(total / trial)

```

```python
plt.plot(trials, frec)
plt.xlabel('Trials')
plt.ylabel('Relative frequency')
plt.axhline(0.5, color='r')
plt.grid(True)
plt.show()
```
